```python
star_wars
```

```python
star_wars_corMatrix = pd.DataFrame(user_id_movie_title_matrix.corrwith(star_wars), columns= ['Correlation'] )
```

### Full movies recommendation system

```python
user_id_movie_title_matrix
```

```python
movie_correlation = user_id_movie_title_matrix.corr(method='pearson', min_periods= 80)
movie_correlation
```

```python
my_rating = pd.read_csv('My_Ratings.csv')
my_rating
```

```python
my_rating['Movie Name'][0]
```

```python
similar_movie_list = pd.Series()
for i in range(len(my_rating)):
  similar_movie = movie_correlation[my_rating['Movie Name'][i]].dropna()
  similar_movie = similar_movie.map(lambda x : x*my_rating['Ratings'][i])
  similar_movie_list = similar_movie_list.append(similar_movie)
```

```python
similar_movie_list.sort_values(inplace= True, ascending=False)
```
